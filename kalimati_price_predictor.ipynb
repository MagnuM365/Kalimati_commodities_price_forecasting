```python
main_df = pd.read_csv('/content/drive/MyDrive/CSV_data/Kalimati_prices.csv')
main_df.head()
```

```python
main_df.info()
```

```python
banana_df = main_df[main_df['Commodity'] == 'Banana']
banana_df
```

```python
banana_df = banana_df[banana_df['Average'] < 200]
```

```python
plt.figure(figsize=(12, 6))
plt.plot('Date', 'Average', data=banana_df)
plt.title('Banana Average Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
from prophet import Prophet
prophet_df = banana_df[['Date', 'Average']]
prophet_df.columns = ['ds', 'y']
prophet_df
```

```python
train = prophet_df[:int(len(prophet_df)*0.8)]
test = prophet_df[int(len(prophet_df)*0.8):]
model = Prophet()
model.fit(train)
```

```python
future = model.make_future_dataframe(periods=2000)
forecast = model.predict(future)
```

```python
forecast[forecast['ds'] == '2025-05-20']
```

```python
actual = test.set_index('ds')['y']
predicted = forecast.set_index('ds').loc[actual.index]['yhat']
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(actual, predicted)
mse = mean_squared_error(actual, predicted)
rmse = np.sqrt(mse)


mape = np.mean(np.abs((actual - predicted) / actual)) * 100
print("Mean Absolute Percentage Error (MAPE):", mape)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
model.plot(forecast)
plt.title('Banana Average Price Forecast')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.show()
```
